```python
import pandas as pd
import os
import glob

folderPath = r"C:\Users\Denys\OneDrive\climatezones"

pattern = os.path.join(folderPath,"*.xlsx")

excelFilePaths = []


currentIndex = 0

for file in glob.glob(pattern):
    if os.path.basename(file)[0:2] != "~$":
        excelFilePaths.append(file)
        
for excelFile in excelFilePaths:
    df_currentFile = pd.read_excel(excelFile, header=None)

    for index, row in df_currentFile.iterrows():
        previousIndex = currentIndex
        if str(row[0])[0:5] == "TABLE":
            currentIndex = index
            print(df_currentFile[previousIndex:currentIndex])
        if row[0] == "(continued)":
            df_currentFile.drop(index,inplace=True)
        if not row.isnull().all():
            continue
        df_currentFile.drop(index,inplace=True)
```
